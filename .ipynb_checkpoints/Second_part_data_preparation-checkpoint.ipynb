# Prepare data

```python
import pandas as pd

data_id = pd.read_csv('kaggle_data/data_identification.csv')
emotion = pd.read_csv('kaggle_data/emotion.csv')
```

```python
import json

file = open("kaggle_data/tweets_DM.json", 'r', encoding='utf-8')
original_data = []
for line in file.readlines():
    dic = json.loads(line)
    original_data.append(dic)

print(type(original_data[0]))
original_data[0]
```

```python
for item in original_data:
    item.update(item['_source']['tweet'])
    del item['_source']
    
original_data[0]
```

```python
original_data = pd.DataFrame(original_data)
original_data
```

# Data analysis and preprocess

## merge dataframe and delete irrelevent feature

```python
data = original_data.copy()
data
```

```python
print(data['_index'].unique())
print(data['_type'].unique())
```

```python
data.drop('_score', axis=1, inplace=True)
data.drop('_index', axis=1, inplace=True)
data.drop('_crawldate', axis=1, inplace=True)
data.drop('_type', axis=1, inplace=True)
data
```

```python
data_id
```

```python
data = data.merge(data_id, on='tweet_id')
data
```

```python
test_data = data[data['identification'] == 'test']
test_data.to_pickle("kaggle_data/test_data.pkl")
test_data
```

```python
emotion
```

```python
train_data = data.merge(emotion, on='tweet_id')
train_data.to_pickle("kaggle_data/train_data.pkl")
train_data
```
